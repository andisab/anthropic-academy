```python
# Verify that model has been enabled in Model Garden. See ./screenshots/VertexAI-setup.png
# Then: connect
from anthropic import AnthropicVertex

client = AnthropicVertex(region="global", project_id="your-project-id")
model = "claude-sonnet-4@20250514"
```

```python
# Make a first request
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)
```

```python
# Multi-turn chat
def add_user_message(messages, text):
    messages.append({"role": "user", "content": text})

def add_assistant_message(messages, text):
    messages.append({"role": "assistant", "content": text})

def chat(messages, user_input, system=None, temperature=0.25): 
    add_user_message(messages, user_input)
    
    params = {
        "model": model, 
        "max_tokens": 1000, 
        "messages": messages,
        "temperature": temperature,
    }
    if system: 
        params["system"] = system

    message = client.messages.create(**params)
    reply = message.content[0].text
    add_assistant_message(messages, reply)
    return reply
```
